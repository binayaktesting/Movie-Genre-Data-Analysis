```python
import pandas as pd
```

```python
##load data
movies=pd.read_csv(r"C:\Users\Binayak\Downloads\imdb_movies.csv")
```

```python
pd.set_option('display.max.rows',11000)
pd.set_option ('display.max.columns',22)
```

```python
## # Checking for duplicate records in the dataset
movies[movies.duplicated()]
```

```python
movies.drop_duplicates(inplace=True) ### remove duplicates
```

```python
## make sure genres and id should be same
movies.dropna(subset=['genres'],inplace=True)
```

```python
movies.head()
```

```python
movies['profit']=movies['revenue']-movies['budget']
```

```python
###get rid of some of the columns
movies_genre=movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
movies_genre.head(5)
```

```python
from pandas import Series, DataFrame
movies_genre['genres'].str.split('|').apply(Series,1).stack()
```

```python
### make the same index for the same string
split= movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index= split.index.droplevel(-1)
split.name='genres_split'
del movies_genre['genres']
movies_genre=movies_genre.join(split)
```

```python
movies_genre
```

###Group by the genre and counting on them

```python
movies_genre.groupby('genres_split').count()
```

```python
##To get the index and sort the values
genres_count=pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending=False)
```

```python
genres_count
```

```python
### Plot a horizontal bar chart showing the number of movies per genre, 
# sorted in descending order so that the most frequent genres appear at the top.
# The chart is styled with a dark blue color and set to a size of 10x9 for better visibility.
genres_count['original_title'].sort_values(ascending=True).plot.barh(
    title='Movies Per Genre', 
    color='darkblue', 
    figsize=(10, 9)
)
```

```python
# Group the dataset by the 'genres_split' column and calculate the mean of all numeric columns.
# The 'numeric_only=True' parameter ensures that only numerical values are averaged, avoiding errors with non-numeric data.
genres_avg = movies_genre.groupby('genres_split').mean(numeric_only=True)
genres_avg
```

```python
genres_avg = movies_genre.groupby('genres_split').mean(numeric_only=True)
pd.options.display.float_format='{:2f}'.format ## change to integer
genres_avg
```

```python
### Plot a horizontal bar chart showing the number of movies per genre, 
# sorted in descending order so that the most frequent genres appear at the top.
# The chart is styled with a dark blue color and set to a size of 10x9 for better visibility.
genres_avg[['budget', 'revenue']].sort_values(by='budget', ascending=True).plot.barh(
    title='Budget and Revenue by Genre', 
    color=['darkblue', 'orange'],  # Adjust colors as needed
    figsize=(10, 9)
)
```

```python
genres_avg
```

```python
genres_avg[['profit']].sort_values(by='profit', ascending=True).plot.barh(
    title='Profit by Genre', 
    color=['darkblue'],  # Adjust colors as needed
    figsize=(10, 9)
)
```

```python
###get rid of some of the columns
movies_genre=movies[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
```

```python
from pandas import Series, DataFrame
```

```python
### make the same index for the same string
split= movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index= split.index.droplevel(-1)
split.name='genres_split'
del movies_genre['genres']
movies_genre=movies_genre.join(split)
```

```python
movies_genre.head(5)
```

```python
## look to profit per genre per year
time_genre =pd.DataFrame(movies_genre.groupby(['release_year','genres_split'])['profit'].mean())
```

```python
time_genre
```

```python
final_genre=pd.pivot_table(time_genre,values ='profit', index =['genres_split'], columns =['release_year'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set figure size
sns.set(rc={'figure.figsize': (15, 10)})

# Generate heatmap
sns.heatmap(final_genre, cmap='YlGnBu', linewidths=0.5)

# Add title
plt.title('Genres by Profit per Year')

# Show the plot
plt.show()
```
